```python
from pandas import read_csv
data = read_csv('data.csv')
```

```python
transactions = data["text_keywords"]
transactions = transactions.str.split(';')
transactions = transactions.apply(lambda x: [str(i) for i in x])
#calculate number of keywords for all transactions
num_keywords = transactions.apply(lambda x: len(x))
num_keywords = num_keywords.sum()
num_keywords_mean = num_keywords/len(transactions)
print(num_keywords)
print(num_keywords_mean)
```

```python
#Create a bar chart of the top 20 frequent itemsets
import matplotlib.pyplot as plt
import pandas as pd

with open('output.txt', 'r') as file:
    lines = file.readlines()

data = []
for line in lines:
    items, count = line.rsplit(' ', 1)
    count = int(count.strip('()\n'))
    data.append((items, count))

df = pd.DataFrame(data, columns=['Itemset', 'Count'])
top_20 = df.sort_values('Count', ascending=False).head(20)

plt.figure(figsize=(15, 10))
bars = plt.bar(range(len(top_20)), top_20['Count'], align='center')
plt.xticks(range(len(top_20)), top_20['Itemset'], rotation=45, ha='right')


plt.title('Top 20 Frequent Itemsets')
plt.xlabel('Itemsets')
plt.ylabel('Count')
plt.tight_layout()

plt.show()
```
